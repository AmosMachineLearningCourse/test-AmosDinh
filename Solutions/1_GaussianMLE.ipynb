Zum Gauß-Klassifikations-Beispiel

```python
import torch
# Sample observations from a normal distribution with mean an variance for A and B

torch.manual_seed(42)  # Wenn wir "random" Zahlen generieren, wollen wir bei jeder Ausführung reproduzierbare Ergebnisse, ein Seed stellt das sicher

sigma_squared_A = 2
mu_A = 4
num_points_A = 12

sigma_squared_B = 3
mu_B = 7
num_points_B = 20

observations_A = torch.normal(mu_A, torch.sqrt(torch.tensor(sigma_squared_A)), size=(num_points_A,))  # Erstelle eine 1D Tensor mit n Observationen aus einer Normalverteilung mit Mittelwert mu_A und Varianz sigma_squared_A
observations_B = torch.normal(mu_B, torch.sqrt(torch.tensor(sigma_squared_B)), size=(num_points_B,))
```

```python
# plotting
import matplotlib.pyplot as plt
plt.plot(observations_A, torch.zeros_like(observations_A), 'x', color='blue', label='A')
plt.plot(observations_B, torch.zeros_like(observations_B), 'o', mfc='none', color='green', label='B')

plt.legend()
plt.show()
```

```python
# Definiere PDF (Probability Density Function) der Normalverteilung/Gaussverteilung
# Dies hier ist eine Funktion
# In python wird die Funktion mit def definiert
# Die Funktion hat die Parameter x, mu und sigma_squared
# Der Rückgabewert ist der Wert der Gaussverteilung an der Stelle x
# Durch Einrücken wird der Codeblock definiert, der zur Funktion gehört und ausgeführt wird, wenn die Funktion aufgerufen wird

# Bspsweise:
def add_and_subtract(a, b):
    x = a + b - 1
    return x

# Aufruf
print(add_and_subtract(2, 3)) # 4

def gaussian(x, mu, sigma_squared):
    return 1/torch.sqrt(2*torch.pi*torch.tensor(sigma_squared)) * torch.exp(-0.5*(x-mu)**2/sigma_squared)
```

```python
plt.plot(observations_A, torch.zeros_like(observations_A), 'x', color='blue', label='A')
plt.plot(observations_B, torch.zeros_like(observations_B), 'o', mfc='none', color='green', label='B')
# Create x values for plotting the distributions
x = torch.linspace(0, 12, 200)

# Plot the probability density functions
plt.plot(x, gaussian(x, mu_A, sigma_squared_A), color='blue', label='PDF A')
plt.plot(x, gaussian(x, mu_B, sigma_squared_B), color='green', label='PDF B')
plt.title('Echte Verteilungen: Die "p_1(x) und p_2(x)" PDFs') 

plt.legend()
plt.show()
```

In der Präsentation haben wir gesehen, dass wir die richtigen Werte für μ und σ² wie folgt berechnen können:

$\mu = \frac{1}{N} \sum_{i=1}^N x_i$ <br><br>
$\sigma^2 = \frac{1}{N} \sum_{i=1}^N (x_i - \mu)^2$

Wobei:
- N die Anzahl der Beobachtungen ist
- x_i die einzelnen Beobachtungen sind


=== Aufgabe === <br>
Berechne μ und σ² für observations_A und obeservations_B (Das ist der Vektor, der unsere x_i enthält).<br>
=== Aufgabe ===  <br><br>
Tipp:
1. Du kannst das Shape Attribut nutzen um "N" zu finden. <br>
Bsp: x.shape[0] # In der Informatik beginnen Aufzählungen bei 0, dieser Ausdruck gibt Shape an der ersten Stelle zurück.
Schau dir observations_A.shape an.
2. In pytorch sind Operationen der Art "Skalar verknüpft Tensor" punktweise definiert: s + T = s + (T)_ij für +,-,*,/ <br>
3. In Pytorch gibt es Aggregatoren: torch.sum(torch.tensor([1,2,3])) = 6, analog torch.mean, torch.max, torch.min (Achtung: Ist ein Tensor multidimensional müssen die Dimensionen angegeben werden, über die aggregiert werden soll. Das kann hier vernachlässigt werden)
4. Auch nützlich, aber nicht notwendig: "b hoch p" = torch.pow(b,p), .pow ist auch Punktweise definiert, b muss ein Tensor sein.


```python
# === Antwort ===
n_A = observations_A.shape[0]
mu_A_hat = 1/n_A * torch.sum(observations_A)
sigma_squared_A_hat = 1/n_A * torch.sum((observations_A - mu_A)**2)

n_B = observations_B.shape[0]
mu_B_hat = 1/n_B * torch.sum(observations_B)
sigma_squared_B_hat = 1/n_B * torch.sum((observations_B - mu_B)**2)
# === Antwort ===
```

Visualisierung der gefundenen q_1(x), q_2(x)

```python
plt.plot(observations_A, torch.zeros_like(observations_A), 'x', color='blue', label='A')
plt.plot(observations_B, torch.zeros_like(observations_B), 'o', mfc='none', color='green', label='B')
# Create x values for plotting the distributions
x = torch.linspace(0, 12, 200)

# Plot the probability density functions
plt.plot(x, gaussian(x, mu_A, sigma_squared_A), color='blue', label='P_A(x)')
plt.plot(x, gaussian(x, mu_B, sigma_squared_B), color='green', label='P_B(x)')
plt.title('Q_A und Q_B sind unsere Estimates der Verteilungen') 

# Plot the estimated distributions
plt.plot(x, gaussian(x, mu_A_hat, sigma_squared_A_hat), '--', color='blue', label='Q_A(x)')
plt.plot(x, gaussian(x, mu_B_hat, sigma_squared_B_hat), '--',  color='green', label='Q_B(x)')

plt.legend()
plt.show()
```

Wir wissen (Bayes Theorem):

$P(A|x) = \frac{P(x|A) \cdot P(A)}{P(x)}$<br>
$Posterior = \frac{Likelihood \cdot Prior}{Evidence}$


=== Aufgabe === <br>
Finde die Entscheidungsregel und somit die Posteriori-Wahrscheinlichkeiten $P(A|x)$ und $P(B|x)$ für den Wert $x = 4.5$ <br>
Wir nehmen an, dass x entweder von den Verteilungen $P_A$ oder $P_B$ stammt.
<br><br>
=== Erläuterung === <br>
<br>
Es gilt:<br><br>
$P(A | x) \varpropto P(x|A) \cdot P(A)$ <br>
$P(B | x) \varpropto P(x|B) \cdot P(B)$ <br>
<br>
Wir ersparen uns die Berechnung von $P(x)$, da diese für A und B gleich ist, und eine aufwendige Integralberechnung benötigen würde.<br>
Somit schreiben wir auf die Frage "Stammt x = 4.5 von $P_A$ oder $P_B$?":<br><br>
$P(A | x) \geq P(B | x) \leftrightarrow P(x|A) \cdot P(A) \geq P(x|B) \cdot P(B)  \leftrightarrow x \sim P_A$

<br><br>
Weiter sagen wir nach Annahme:<br><br>
$P(A | x) + P(B | x) = 1$ 

<br>und somit<br><br>
$ P(A | x) = \frac{P(x|A) \cdot P(A)}{P(x|A) \cdot P(A) + P(x|B) \cdot P(B)} $ <br><br>
$ P(B | x) = 1 - P(A | x)$







 <br> <br>
Tipps: 
1. Du musst die Formulierungen für alle Wahrscheinlichkeiten auf der rechten Seite der ersten Gleichung in der Erläuterung finden. 
2. Wofür stehen "gaussian(x, mu_A, sigma_squared_A)", und "gaussian(x, mu_A, sigma_squared_A)", welche wir weiter oben definiert haben? Nutze sie.


```python
# Weiter gehts

def posterior_A(x, observations_A, observations_B, mu_A, sigma_squared_A, mu_B, sigma_squared_B):
    # === Aufgabe ===

    # Anzahl der Beobachtungen
    N_A = observations_A.shape[0]
    N_B = observations_B.shape[0]

    # Priors 
    p_A = N_A / (N_A + N_B)
    p_B = N_B / (N_A + N_B)

    # Likelihoods
    p_x_given_A = gaussian(x, mu_A, sigma_squared_A)
    p_x_given_B = gaussian(x, mu_B, sigma_squared_B)


    # Evidence ignorieren wir, wird in der Praxis mit N Stichproben von mu und sigma^2 approximiert wobei dann p(x) = 1/N * sum p(x|mu_i, sigma^2_i)
    # p_x = "integral integral über alle Werte von mu und sigma^2 von p(x|mu, sigma^2) * p(mu, sigma^2) dmu dsigma^2"

    # "P(A|x)"
    # Wahrscheinlichkeit das A x generiert hat unter der Annahme, dass x nur von A oder B generiert werden kann:
    p_A_given_x = p_A * p_x_given_A / (p_A * p_x_given_A + p_B * p_x_given_B)
    
    # === Aufgabe ===
    return p_A_given_x

x = torch.tensor(4.5)

post_A = posterior_A(x, observations_A, observations_B, mu_A, sigma_squared_A, mu_B, sigma_squared_B)
print('P(A|x) =', post_A)
```

Visualisierung

```python
# Create x values for plotting
x = torch.linspace(0, 10, 200)

# Create figure with three subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 15))

# Top plot - distributions and observations
ax1.plot(observations_A, torch.zeros_like(observations_A), 'x', color='blue', label='A')
ax1.plot(observations_B, torch.zeros_like(observations_B), 'o', mfc='none', color='green', label='B')
ax1.plot(x, gaussian(x, mu_A, sigma_squared_A), color='blue', label='P_A(x)')
ax1.plot(x, gaussian(x, mu_B, sigma_squared_B), color='green', label='P_B(x)')
ax1.plot(x, gaussian(x, mu_A_hat, sigma_squared_A_hat), '--', color='blue', label='Q_A(x)')
ax1.plot(x, gaussian(x, mu_B_hat, sigma_squared_B_hat), '--', color='green', label='Q_B(x)')
ax1.set_title('Distributions and Observations')
ax1.legend()


# Bottom plot - posterior probabilities
posteriors_A = [posterior_A(xi, observations_A, observations_B, mu_A, sigma_squared_A, mu_B, sigma_squared_B) for xi in x]
posteriors_B = [1 - p_A for p_A in posteriors_A]  # P(B|x) = 1 - P(A|x) since they are complementary
ax2.plot(x, posteriors_A, color='blue',  linestyle='--', label='P(A|x)')
ax2.plot(x, posteriors_B, color='green', linestyle='--', label='P(B|x)') 
ax2.axhline(y=0.5, color='k', linestyle='--', alpha=0.5)
ax2.axvline(x=4.5, color='red', linestyle='-', alpha=0.5)  # Add vertical line at x=4.5
ax2.set_title('Posterior Probabilities P(A|x) and P(B|x)')
ax2.set_xlabel('x')
ax2.set_ylabel('Probability')
ax2.legend()

plt.tight_layout()
plt.show()
```

Warum nutzen wir Likelihood und Prior?<br> 
Intuition:<br>
p(x|A) gibt uns die Wahrscheinlichkeit dass x ~ p_A. D.h. Ein "Update" von dem was wir glauben, basierend auf dem Wert von x.<br>
p(A) ist die Apriori-Wahrscheinlichkeit, dass irgendein Datenpunkt von A stammt und nicht von B.<br>
Das ist "Bayesian Inference" und besonders bei geringen Datenmengen in der Praxis robuster als nur p(x|A) "Frequentist Approach" zu verwenden:


Wir haben 12 Datenpunkte von A und 20 von B gesammelt und sehen im zweiten Plot (Neue Entscheidungsgrenze) im Vergleich zum ersten (likelihood), dass sich der Schnittpunkt der beiden Funktionen nach links verschiebt. Das liegt daran, dass wir 8 mehr Datenpunkte von B haben und somit durch den Prior encodieren dass wir uns bei dem Aussehen der rechten Verteilung "sicherer" sind. <br>
Bemerkung: Der Schnittpunkt der Likelihood Funktionen wäre unsere Entscheidungsgrenze im Fall, dass wir nur  p(x|C) berücksichtigen.
